```python
#import the libs

import pandas as pd
import numpy as np
import matplotlib.patches as mpaches
import matplotlib as mpl
import backtrader as bt 
import yfinance as yf
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')


from termcolor import colored as cl # Color formatting for output in terminal

from datetime import datetime

```

```python
stock = 'BA'
startdate = '2021-08-10'
enddate = '2022-08-11'


df = yf.download(stock,start=startdate, end=enddate,progress=False)
df.head()
```

```python
df = df.loc[:,['Adj Close']]
df.rename(columns={'Adj Close':'adj_close'}, inplace=True)
df['simple_return']=df.adj_close.pct_change() 
df['log_return']=np.log(df.adj_close/df.adj_close.shift(1))
df.head()
```

```python
fig,ax= plt.subplots(3,1 ,figsize=(24,30),sharex=False)
plt.subplots_adjust(bottom=0.1,  top=0.9, hspace=0.4)

df.adj_close.plot(ax=ax[0],fontsize=16)

ax[0].set_title('Boeing Close price history',fontsize=20,color='brown')
ax[0].set_ylabel('close price (USD)',fontsize=20,color='brown')
ax[0].set_xlabel('Date',fontsize=20,color='brown')


df.simple_return.plot(ax=ax[1],fontsize=16)
ax[1].set_title('Boeing stock simple return',fontsize=20,color='brown')
ax[1].set_ylabel('Simple return (%)',fontsize=20,color='brown')
ax[1].set_xlabel('Date',fontsize=20,color='brown')

df.log_return.plot(ax=ax[2],fontsize=16)
ax[2].set_title('Boeing stock log return',fontsize=20,color='brown')
ax[2].set_ylabel('Log return (%)',fontsize=20,color='brown')
ax[2].set_xlabel('Date',fontsize=20,color='brown')
```

```python
cash = 10000000
commission = 0.001
percentageUsed = 10

prev_cash = cash

class SMAStrategy (bt.Strategy):    
    params = (('p1', 10), )
    
    
    def __init__ (self):
        self.order = 0
        self.price = 0
        self.comm = 0
        self.sma = bt.ind.SimpleMovingAverage(self.data0,period=self.params.p1)
        

    def log(self, notification):
        dt=self.datas[0].datetime.date(0).isoformat()
        print(f'{dt} | {notification}')

    def notify_order(self, order):
        if order.status in [order.Submitted, order.Accepted]:
            return
        
        if order.status in [order.Completed]:
            if order.isbuy():
                self.log(cl(f'   BUY ORDER EXECUTED!  [Price:{order.executed.price:.2f} - Cost:{order.executed.value:.2f} - Commission:{order.executed.comm:.2f}]',color='green',attrs=['bold']))
                self.price = order.executed.price
                self.comm = order.executed.comm 
            else:
                self.log(cl(f'   SELL ORDER EXECUTED! [Price:{order.executed.price:.2f} - Cost:{order.executed.value:.2f} - Commission:{order.executed.comm:.2f}]',color='red',attrs=['bold']))        
                self.bar_executed = len(self)

        elif order.status in [order.Canceled, order.Margin, order.Rejected]:
                    self.log(cl('â˜“ Order Failed!',color='red'))
        self.order=None
        
    
    def notify_trade(self, trade):
        self.log(cl(f'OPERATION RESULTS: [GROSS PNL: {trade.pnl:.2f} - NET PNL:{trade.pnlcomm:.2f}]',attrs=['bold']))
    
    def notify_fund(self,cash, value, fundvalue, shares):
        global prev_cash
        if(cash == prev_cash):
            return
        else:
            self.log(cl(f'CURRENT FUNDS:   [CASH: {cash:.2f} - VALUE:{value:.2f} - FUND VALUE:{fundvalue:.2f} - SHARES:{shares:.2f}]',attrs=['bold']))
            prev_cash = cash
        
#     
    def next(self):
#             
        if self.position:
            if self.datas[0].close[0] < self.sma[0]:
                    self.log(cl(f'â–¶ SELL ORDER IS CREATED! Price:{self.datas[0].close[0]:.2f}',color='red',attrs=['bold']))
                    self.order = self.sell()
            
        else:
            if self.datas[0].close[0] > self.sma[0]:
                self.log(cl(f'â–¶ BUY ORDER IS CREATED!   Price:{self.datas[0].close[0]:.2f}',color='green',attrs=['bold']))
                self.order = self.buy()
                    
```

```python
def uselessFunction(uselessArg):
    print('\n')
    print( '+' * 50)
    print('+',' '*46,'+')
    print ('+',' '*4,uselessArg,' '*4,)
    print('+',' '*46,'+')
    print('+'*50,'\n')
plt.figure().clear()
```

```python

cerebro = bt.Cerebro(stdstats=False)
data = bt.feeds.PandasData(dataname=yf.download(stock,startdate ,enddate ))

cerebro.adddata(data)
cerebro.broker.setcash(cash)   
cerebro.addstrategy(SMAStrategy)
cerebro.broker.setcommission(commission)
cerebro.addsizer(bt.sizers.PercentSizer,percents=percentageUsed)  
cerebro.addobserver(bt.observers.BuySell)
cerebro.addobserver(bt.observers.Value)
uselessFunction(f'Our Portfolio Value: {cerebro.broker.getvalue()}')

cerebro.broker.getvalue()
cerebro.run()
print()
uselessFunction(f'Final Portfolio Value:  {cerebro.broker.getvalue():.2f}')
uselessFunction(f'Profit is : {cerebro.broker.getvalue() - cash:.2f}')

plt.rcParams['figure.figsize'] = [6, 6]
plt.style.use('seaborn-darkgrid')

cerebro.plot()
```
